### Ejercicio 1

Representa en un único gráfico las funciones arcoseno y arcocoseno en el intervalo [-pi/4, pi/4]  (en radianes):

```python
%matplotlib inline
import matplotlib
import math
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-np.pi/4, np.pi/4,0.1)
plt.plot(x,np.arcsin(x),'b',x,np.arccos(x),'y')
plt.xlabel('x')
plt.ylabel('azul: arcoseno(x), amarillo:arcocoseno(x)')
```

### Ejercicio 2

Representa en un gráfico 3D la función `f(x,y) = cos(x) + 2*sen(y)` en el intervalo [-2*pi, 2*pi] (en radianes):

```python
%matplotlib inline

# Función f(x,y) = cos(x) + 2*sen(y)
# Importamos paquete 3D
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
import math
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-2*np.pi,2*np.pi,0.1)
y=np.arange(-2*np.pi,2*np.pi,0.1)
grafica = plt.figure()
ax= grafica.gca(projection='3d')
ax.plot(x,y,np.cos(x)+2*np.sin(y), 'r')
```

### Ejercicio 4

Dados los siguientes datos de meteorología registrados en el aeropuerto de Barcelona, representa en uno o varios gráficos los valores de temperatura media anual, máxima, mínima, media de lluvias mensual y humedad relativa.

Las posiciones del vector `meteo` son las siguientes:

* Nombre mes
* Temperatura media anual del mes (°C)
* Temperatura máxima media (°C)
* Temperatura mínima media (°C)
* Media de lluvias mensual (mm)
* Humedad media relativa (%)


```python
%matplotlib inline
import matplotlib
import math
import numpy as np
import matplotlib.pyplot as plt
meteo = [["January", 9.2, 13.6, 4.7, 37, 70],
["February",9.9, 14.3, 5.4, 35, 70],
["March", 11.8, 16.1, 7.4, 36, 70],
["April", 13.7, 18.0, 9.4, 40, 69],
["May", 16.9, 21.1, 12.8, 47, 70],
["June", 20.9, 24.9, 16.8, 30, 68],
["July", 23.9, 28.0, 19.8, 21, 67],
["August", 24.4, 28.5, 20.2, 62, 68],
["September", 21.7, 26.0, 17.4, 81, 70],
["October", 17.8, 22.1, 13.5, 91, 73],
["November", 13.0, 17.3, 8.6, 59, 71],
["December", 10.0, 14.3, 5.7, 40, 69]]
# Creamos un array con todos los datos 
x=np.array(meteo)
# Ahora los componentes del array son un string por lo que los cambiamos a float 
# Asignamos a cada columna una variable
meses=x[:,0]
tmedia=x[:,1].astype(np.float)
tmax=x[:,2].astype(np.float)
tmin=x[:,3].astype(np.float)
lluvias=x[:,4].astype(np.float)
humedad=x[:,5].astype(np.float)

# Seleccionamos una grafica de matplotlib que nos permita representar estas variables que tienen distintas variables
fig,host=plt.subplots()
fig.subplots_adjust(right=4,top=2.5)
# Añadimos este comando para generar más ejes Y para representar las 3 variables
par1=host.twinx()
par2=host.twinx()
par2.spines["right"].set_position(("axes", 1.1))
par2.spines["right"].set_visible(True)

# Ahora representamos los valores
# Los tres primeros representan temperatura
var1, = host.plot(meses,tmedia, "p", label="T media anual del mes")
var2, = host.plot(meses,tmax, "g", label="T máxima media")
var3, = host.plot(meses, tmin, "y", label="T mínima media")
# Para estos usamos los nuevos ejes
var4, = par1.plot(meses,lluvias, "b", label="Media de lluvias")
var5, = par2.plot(meses, humedad, "r", label="Humedad relativa")

host.set_ylim(0,50)
par1.set_ylim(0, 100)
par2.set_ylim(50,100)

# Etiquetas
host.set_xlabel("Meses")
host.set_ylabel("T(ºC)")
par1.set_ylabel("Lluvia(mm)")
par2.set_ylabel("Humedad(%)")

tkw = dict(size=5, width=7)
lines = [var1, var2, var3, var4, var5]

host.legend(lines, [l.get_label() for l in lines])
```
